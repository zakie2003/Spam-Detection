```python
import xml.etree.cElementTree as et
import pandas as pd
```

```python
tree=et.parse('./DataSet/sms-20240825182559.xml')
root=tree.getroot()
```

```python
address=[]
body=[]
```

```python
sms_data = []

    # Iterate over each SMS entry
for sms in root.findall('sms'):
        sms_info = {
            'address': sms.get('address'),
            'body': sms.get('body')
        }
        sms_data.append(sms_info)
    
sms_data
```

```python
for i in sms_data:
    address.append(i["address"])
    body.append(i["body"])
```

```python
pd.DataFrame({"Address":address,"Body":body}).to_csv("./DataSet/sms_spam.csv")
```
